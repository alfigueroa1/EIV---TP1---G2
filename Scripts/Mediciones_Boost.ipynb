```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from utils import *

# Folder path
path = '..\Mediciones\\'
```

```python
def plot_measurement(path, filename):
	# Create dataframe
	column_names = ['t', 'V']  # specify column names
	df = pd.read_csv(path + filename + '.csv', header=None, skiprows=2, names=column_names)  # skip the first row, but use column_names as column names

	# Load the image
	img = mpimg.imread(path + filename + '.png')

	# Show the image and the graph side by side
	fig, ax = plt.subplots(1, 2, figsize=(14, 4))
	fig.suptitle(filename)
	_, exp = get_exp(df['t'].abs().max())
	ax[0].set_xlabel('t [' + get_unit_prefix(df['t'].abs().max()) + 's]')
	ax[0].set_ylabel('V [V]')
	ax[0].plot(df['t'].apply(lambda t: t/(1E3)**exp), df['V'])
	ax[0].grid()
	ax[1].imshow(img)
	ax[1].set_axis_off()

	return ax
```

### $V_o$ CCM

```python
filename = 'Vo_CCM'
ax = plot_measurement(path, filename)
ax[0].set_ylim([-0.1, 14])
plt.savefig('..\Grficos\\' + filename + '.png', dpi=300)
plt.show()
```

### $V_o$ CCM (Zoom)

```python
filename = 'Vo_CCM_zoom'
ax = plot_measurement(path, filename)
ax[0].set_ylim([11.5, 13.8])
plt.savefig('..\Grficos\\' + filename + '.png', dpi=300)
plt.show()
```
